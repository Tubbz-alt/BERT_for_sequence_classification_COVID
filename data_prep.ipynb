```python
import pandas as pd
from tqdm import tqdm_notebook

dir = 'data/'
```

```python
train_df = pd.read_csv(dir + 'train.tsv' ,sep="\t")
train_df.head()
```

```python
test_df = pd.read_csv(dir + 'valid.tsv',sep="\t", header = None)
test_df.head()
```

```python
train_df['Label'] = (train_df['Label'] == "INFORMATIVE").astype(int)
test_df[2] = (test_df[2] == "INFORMATIVE").astype(int)
```

```python
test_df.head()
```

```python
train_df = pd.DataFrame({
    'id':range(len(train_df)),
    'label':train_df['Label'],
    'alpha':['a']*train_df.shape[0],
    'text': train_df['Text'].replace(r'\n', ' ', regex=True)
})

train_df.head()
```

```python
dev_df = pd.DataFrame({
    'id':range(len(test_df)),
    'label':test_df[2],
    'alpha':['a']*test_df.shape[0],
    'text': test_df[1].replace(r'\n', ' ', regex=True)
})

dev_df.head()
```

```python
train_df.to_csv('data/train_processed.tsv', sep='\t', index=False, header=False, columns=train_df.columns)
dev_df.to_csv('data/valid_processed.tsv', sep='\t', index=False, header=False, columns=dev_df.columns)
```
